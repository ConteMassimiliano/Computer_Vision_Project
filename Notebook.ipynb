```python
import numpy as np
import matplotlib.pyplot as plt
```

## Load data

```python
X = np.load("data/images.npy")
y = np.load("data/labels.npy")
y = [val[0] for val in y]
```

```python
import seaborn as sns

# seaborn histogram
sns.distplot(y, hist=True, kde=False, 
             bins=40, color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.axvline(np.mean(y), linewidth=3, color='red', label="mean", alpha=0.5)
plt.axvline(np.median(y), linewidth=4, color='green', label="median", alpha=0.5)
# Add labels
plt.legend(["mean", "median"])
plt.title('Distribution of y values in Mall Dataset')
plt.xlabel('Humans in image')
```

```python
to_plot = 9
ncol = 3
plt.figure(figsize=(10, 10))
plt.rcParams['figure.facecolor'] = 'black'

for i in range(to_plot):
    plt.subplot(int(np.floor(to_plot / ncol)) + 1, ncol, i + 1)
    plt.title("Number: " + str(y[i]))
    plt.axis('off')
    _ = plt.imshow(X[i])
```
