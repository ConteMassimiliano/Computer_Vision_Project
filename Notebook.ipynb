```python
import numpy as np
import matplotlib.pyplot as plt
```

## Load data

```python
X = np.load("images.npy")
y = np.load("labels.npy")
y = [val[0] for val in y]
```

```python
print(f"Images are {X[0].shape[0]} x {X[0].shape[1]} pixels with {X[0].shape[2]} color channels")
```

```python
print(f"There are {len(X)} images")
```

```python
print(f"There are a mean of {np.round(np.mean(y), 2)} people per image")
```

```python
to_plot = 9
ncol = 3
plt.figure(figsize=(10, 10))
plt.rcParams['figure.facecolor'] = 'black'

for i in range(to_plot):
    plt.subplot(int(np.floor(to_plot / ncol)) + 1, ncol, i + 1)
    plt.title("Number: " + str(y[i]))
    plt.axis('off')
    _ = plt.imshow(X[i])
```
